1.What is a random variable in probability theory?

  -> In probability theory, a random variable is essentially a mathematical  
     tool used to quantify the outcomes of random processes. Instead of describing outcomes in words (like "rolling a die and getting six"), a random variable assigns numerical values to these outcomes, making them easier to analyze mathematically.



     

2.What are the types of random variables?

  -> Random variables come in two types:

  1.Discrete Random Variables: These take distinct, countable values (e.g., the number on a rolled die: 1, 2, 3, 4, 5, or 6).

  2.Continuous Random Variables: These take values from a continuous range or interval





3.What is the difference between discrete and continuous distributions?

  -> The key difference between discrete and continuous distributions lies in
     the type of values their random variables can take:

     Discrete Distributions:

    Deal with discrete random variables, which take on a countable set of values.

    Examples include rolling a die, where the outcomes are distinct numbers (1, 2, 3, 4, 5, 6), or counting the number of cars passing through a toll booth.

    Probability is assigned to specific values, and the sum of probabilities equals 1.

    Example: Binomial distribution, Poisson distribution.

Continuous Distributions:

    Deal with continuous random variables, which can take any value within an interval or range.

    Examples include measuring a person's height or the amount of rainfall in a day.

    Instead of assigning probabilities to specific values, probabilities are calculated for intervals using a probability density function (PDF).

    Example: Normal distribution, Exponential distribution.




4.What are probability distribution functions (PDF)?

  -> A Probability Distribution Function (PDF) describes the likelihood of a random variable taking on specific values within its range. It serves as a mathematical representation of how probabilities are distributed across possible outcomes.





5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

  -> The main difference between Cumulative Distribution Functions (CDF) and Probability Distribution Functions (PDF) lies in how they represent probabilities:

1.PDF (Probability Distribution Function):

    Shows the likelihood or "density" of a random variable taking on a specific value (for discrete variables) or within a small interval (for continuous variables).

    For a discrete random variable, it is often called the Probability Mass Function (PMF), which assigns probabilities to individual values.

    For a continuous random variable, it represents the curve, and probabilities are calculated as the area under the curve across intervals.

    Example: The bell-shaped curve of a normal distribution.

2.CDF (Cumulative Distribution Function):

    Represents the cumulative probability that a random variable is less than or equal to a certain value.

    It is the running total of probabilities from the smallest possible value up to the given value.

    For a continuous random variable, the CDF is obtained by integrating the PDF up to the specified point.




6.What is a discrete uniform distribution?

  -> A discrete uniform distribution is a type of probability distribution in which all possible values of a discrete random variable are equally likely. In other words, each outcome has the same probability.



7.What are the key properties of a Bernoulli distribution?

  -> 1. Binary Outcomes: The random variable X takes only two possible values:

    X = 1, representing sucess.
    X = 0, representing failure.

  2.Probability of Success: Denoted as p, where 0 ≤ p ≤ 1.This is the probability of X= 1.

  3.Probability of Failure: Denoted as 1-p.This is the probability of  X= 0.

  4.Mean: The expected value of X is:

                E[X] = p

  This represents the average outcome over many trials.


  5.Variance: The measure of variability in outcomes is:

              Var(X) = p (1-p)

   The variance is highest when p = 0.5 (equal chance for success and failure).



8.What is the binomial distribution, and how is it used in probability?

  -> The binomial distribution is a probability distribution that models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (like success/failure, yes/no, or head/tail). It is an extension of the Bernoulli distribution, which deals with just one trial, to multiple trials.



9.What is the Poisson distribution and where is it applied?

  -> The Poisson distribution is a probability distribution that models the
     likelihood of a certain number of events occurring within a fixed interval of time, space, or other measurable dimension, given the events happen independently and at a constant average rate

     uses: The Poisson distribution is widely used when you are counting
           occurrences of events over a fixed interval and those events occur randomly but independently:

          - Customer Service: Predicting the number of calls a call center
            receives per hour.

          - Traffic Management: Estimating the number of cars arriving at       toll booth in a given period.



10.What is a continuous uniform distribution?

  ->The continuous uniform distribution is a probability distribution in which all values within a certain interval are equally likely to occur. It's often referred to as a "rectangular distribution" because its probability density function (PDF) is constant over the interval, forming a flat, horizontal line



11.What are the characteristics of a normal distribution?

  -> The normal distribution, often called the "bell curve," is one of the most important probability distributions in statistics. Here are its key characteristics:

Key Features:
1.Symmetry:
The normal distribution is symmetric around its mean (𝜇).

The left side is a mirror image of the right side.

2.Bell Shape:

The curve is highest at the mean and gradually decreases as you move away from it.

Most of the data is concentrated near the mean.

3.Mean, Median, and Mode:

These three measures of central tendency are all equal in a normal distribution.

4.Standard Deviation (𝜎):

Controls the "spread" or width of the distribution.

A smaller 𝜎 results in a narrower curve, while a larger 𝜎 produces a wider curve.

5.Probability Density Function (PDF):

The formula is:

\[ f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e{-\frac{(x-\mu)2}{2\sigma^2}} \]

𝜇: Mean (center of the distribution).

𝜎: Standard deviation.

6.Empirical Rule:

Approximately 68% of values fall within one standard deviation (𝜇±𝜎).

95% within two standard deviations (𝜇±2𝜎).

99.7% within three standard deviations (𝜇±3𝜎).

13.What is the Central Limit Theorem (CLT), and why is it critical in
   statistics?

  -> The Central Limit Theorem (CLT) is a fundamental principle in statistics
     that describes the behavior of sample means for large samples, regardless of the original distribution of the population.

  Why It’s Critical:
  Foundation of Inferential Statistics:

  The CLT allows statisticians to make predictions and draw conclusions about a population based on sample data.

  It justifies the use of normal distribution-based methods, such as hypothesis testing and confidence intervals.



15.What is the application of Z statistics in hypothesis testing?

  -> Applications of Z-Statistics in Hypothesis Testing:

Testing Means:

When the sample size is large (𝑛>30), Z-statistics are used to test whether the sample mean differs significantly from a population mean.

Example: Comparing the average height of a group of people to the general population average.

Proportion Testing:

Z-statistics are used to test hypotheses about population proportions.

Example: Checking if the proportion of customers who prefer a product matches a claimed proportion.

Two-Sample Testing:

Z-statistics can compare the means of two independent large samples.

Example: Evaluating if two groups of students scored differently on a test.



16.How do you calculate a Z-score, and what does it represent?

  -> Calculation:

The formula for calculating a Z-score is:  Z = 𝑋−𝜇/𝜎

Where:

𝑋 = the value you're analyzing.

𝜇 = the mean of the dataset.

𝜎 = the standard deviation of the dataset.

Standardized Distance:

  *The Z-score tells you how many standard deviations the value is from the
   mean.

  *A positive Z-score means the value is above the mean.

  *A negative Z-score means the value is below the mean.

Comparison Across Datasets:

  *Z-scores allow you to compare values from different datasets, even if the datasets have different scales or units.

Detection of Outliers:

  *Values with a Z-score greater than +3 or less than -3 are often considered outliers because they are far from the mean.

19.What is the relationship between a Z-score and a confidence interval?

21.What are the assumptions for applying the Central Limit Theorem?

  -> Assumptions for Applying CLT:
1.Independent Observations:

The samples must be independent of each other, meaning the outcome of one observation does not influence another.

2.Random Sampling:

The samples should be randomly selected from the population to avoid bias.

3.Sample Size:

The sample size should be sufficiently large for the CLT to hold.

While there's no fixed number, a sample size of 𝑛≥30
 is often considered "large" enough for most populations. For highly skewed distributions, a larger sample may be needed.

4.Finite Variance:

The population must have a finite variance (𝜎2), meaning the data isn’t excessively spread out.



22.What is the concept of expected value in a probability distribution?

  ->The expected value is a fundamental concept in probability and statistics that represents the average outcome you would expect from a random variable over many repetitions of the experiment. It's essentially a measure of the "center" or "long-term average" of the probability distribution.

Definition:
The expected value 𝐸(𝑋) of a random variable 𝑋 is calculated as the weighted average of all possible values, where the weights are the probabilities of those values.



23.How does a probability distribution relate to the expected outcome of a random variable?


  -> A probability distribution is directly linked to the expected outcome (or
     expected value) of a random variable because it defines how probabilities are assigned to the variable's possible values. Essentially, the probability distribution provides the foundation for calculating the expected value.

Relationship:
Probability Weights:

Each possible value of the random variable has a corresponding probability, as determined by the distribution.

These probabilities serve as "weights" when calculating the expected value, ensuring that more likely outcomes contribute more to the result.
